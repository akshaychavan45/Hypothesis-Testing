# One tail

```python
from scipy import stats
import pandas as pd
data = pd.Series([0.593,0.142,0.329,0.691,0.231,0.519,0.392,0.418,0.793])
```

```python
stats.ttest_1samp(data,0.3)[1]
```

```python
p = stats.ttest_1samp(data,0.3)[1]
p_value = p/2
p_value
```

# Two Tail test

```python
control = pd.Series([91,87,99,77,88,91])
Treat = pd.Series([101,110,103,93,99,104])
```

```python
stats.ttest_ind(control,Treat)
```

```python
Resto_1 = pd.Series([5,3.4,4,5,4.2,5,3.8,3.8,4.2,5])
Resto_2 = pd.Series([4,3.5,5,1,2.5,3,4,3.1,5,4,2,4])
stats.ttest_ind(Resto_1,Resto_2)
```
